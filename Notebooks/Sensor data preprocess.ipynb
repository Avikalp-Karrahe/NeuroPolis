```python
import pandas as pd
import numpy as np

# 1. Load Sensor Data and Convert Timestamp to datetime
sensor_df = pd.read_csv("sensor_readings.csv")
sensor_df['timestamp'] = pd.to_datetime(sensor_df['timestamp'])

# 2. Remove duplicate rows
sensor_df = sensor_df.drop_duplicates()
sensor_df.head()
```

```python
# 3. Define acceptable reading ranges for each sensor type
thresholds = {
    'temp': {'min': -30, 'max': 500},
    'humidity': {'min': 0, 'max': 500},
    'seismic': {'min': 0, 'max': 500},
    'flood': {'min': 0, 'max': 500},
    'air_quality': {'min': 0, 'max': 500},
}

# Function to check if a reading is valid
def is_valid_reading(row):
    sensor_type = row['sensor_type'].lower().strip()  # Normalize sensor type
    reading = row['reading_value']
    if sensor_type in thresholds:
        return thresholds[sensor_type]['min'] <= reading <= thresholds[sensor_type]['max']
    # For sensor types without defined thresholds, assume valid
    return True

# 4. Validate readings and add a helper column
sensor_df['valid_reading'] = sensor_df.apply(is_valid_reading, axis=1)

# 5. Filter out rows with invalid readings or that are faulty
cleaned_sensor_df = sensor_df[
    (sensor_df['valid_reading'] == True) &
    (sensor_df['status'].str.lower() != 'faulty')
].copy()

# 6. Keep both 'active' and 'inactive' sensors using isin() to filter the status
cleaned_sensor_df = cleaned_sensor_df[
    cleaned_sensor_df['status'].str.lower().isin(['active', 'inactive'])
]

# Remove the helper column used for validation
cleaned_sensor_df.drop(columns=['valid_reading'], inplace=True)

# Reset index for cleaner results
cleaned_sensor_df.reset_index(drop=True, inplace=True)

# Print a summary of the cleaned sensor data
print("Cleaned sensor data shape:", cleaned_sensor_df.shape)
display(cleaned_sensor_df.head())

```

```python
from google.colab import files

# Save the cleaned sensor dataframe to a CSV file
cleaned_file_path = "cleaned_sensor_data.csv"
cleaned_sensor_df.to_csv(cleaned_file_path, index=False)

# Download the CSV file
files.download(cleaned_file_path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming cleaned_sensor_df is already created from your previous cleaning step.
# If not, load and clean your sensor data first.

# 1. Overall Summary Statistics and Count Frequencies
print("=== Overall Summary Statistics ===")
display(cleaned_sensor_df.describe())

print("=== Frequency Counts for Categorical Columns ===")
print("Sensor Types:\n", cleaned_sensor_df['sensor_type'].value_counts())
print("\nStatus Counts:\n", cleaned_sensor_df['status'].value_counts())

```

```python
# 2. Box Plot: Sensor reading distributions by sensor type
plt.figure(figsize=(12, 6))
sns.boxplot(x='sensor_type', y='reading_value', data=cleaned_sensor_df)
plt.title("Sensor Reading Distribution by Sensor Type (Cleaned Data)")
plt.xlabel("Sensor Type")
plt.ylabel("Reading Value")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# 3. Histograms: One for each sensor type
unique_sensor_types = cleaned_sensor_df['sensor_type'].unique()
for sensor in unique_sensor_types:
    plt.figure(figsize=(8, 4))
    subset = cleaned_sensor_df[cleaned_sensor_df['sensor_type'] == sensor]
    sns.histplot(subset['reading_value'], bins=30, kde=True)
    plt.title(f"Histogram of Readings for Sensor Type: {sensor}")
    plt.xlabel("Reading Value")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point

# Load the city map (GeoJSON file)
city_map = gpd.read_file("city_map.geojson")

# Create a GeoDataFrame from your cleaned sensor DataFrame using their longitude/latitude
geometry = [Point(xy) for xy in zip(cleaned_sensor_df.longitude, cleaned_sensor_df.latitude)]
sensor_gdf = gpd.GeoDataFrame(cleaned_sensor_df, geometry=geometry, crs="EPSG:4326")

# If your city's map projection is different, re-project sensor_gdf and city_map accordingly.
# For example, if the city_map is in a projected CRS (e.g., EPSG:3857), then:
# sensor_gdf = sensor_gdf.to_crs(city_map.crs)

# Plot the city map and overlay the sensor locations
fig, ax = plt.subplots(figsize=(12, 10))
city_map.plot(ax=ax, color="lightgray", edgecolor="black", alpha=0.5)
sensor_gdf.plot(ax=ax, column="sensor_type", categorical=True, legend=True, markersize=3, alpha=0.8)

plt.title("Geospatial Distribution of Sensors Overlayed on City Map")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.tight_layout()
plt.show()
```

```python
# 5. Time Series Analysis: Hourly count of sensor readings by sensor type.
# Convert timestamp to datetime if not already, and ensure it's the DataFrame index
temp_df = cleaned_sensor_df.copy()
temp_df['timestamp'] = pd.to_datetime(temp_df['timestamp'])
temp_df = temp_df.set_index('timestamp')

# Group by hour and sensor type, count the occurrences (number of readings)
sensor_counts_time = temp_df.groupby([pd.Grouper(freq='H'), 'sensor_type']).size().reset_index(name='counts')

plt.figure(figsize=(12, 6))
sns.lineplot(data=sensor_counts_time, x='timestamp', y='counts', hue='sensor_type')
plt.title("Hourly Count of Sensor Readings by Sensor Type")
plt.xlabel("Timestamp (Hourly)")
plt.ylabel("Reading Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Reset index if needed for further processing
temp_df.reset_index(inplace=True)
```
